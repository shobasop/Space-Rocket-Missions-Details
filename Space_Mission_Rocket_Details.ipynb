#### Mathematical Analysis

```python
total_missions= ds['count'].sum()
ds['percentage'] = (ds['count'])/ total_missions * 100
ds['percentage'] = ds['percentage'].round(2)
print(ds)
```

```python
plt.pie(ds['count'], labels=ds['status_mission'],autopct='%1.1f%%')
plt.title('Mission Status Distribution')
plt.show()
```
